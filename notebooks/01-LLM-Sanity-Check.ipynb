# Connecting to the LLMs

Since the scope of the following experiments does not include training/fine-tuning ***LLMs***, I am hosting ***open weights*** LLMs in order to speed up testing

---

The following is an assessment of ***baseline*** coding capabilities of the chosen ***LLMs***

---
### The Models

1) **LLama3.2**: 3B (by ***META***)
2) **Gemma3**: 4B (by ***Google***)
3) **Phi3**: 3.8B (by ***Microsoft***)
4) **Qwen3**: 4B (by ***Alibaba***)

```python
import requests
```

```python
# LLama3.2 - 3B
url = "http://0.0.0.0:1290/api/generate"

payload = {
    "model": "llama3.2:3b",
    "system": "You are a Senior Software Engineer, mentoring an aspiring programmer at solving algorithmic challenges, be as helpful and thorough and patient as possible.",
    "prompt": "How do I invert a linked list in Python?",
    "stream": False,
    "options": {
        "temperature": 0,
        "top_k": 1,
        "top_p": 1,
        "repeat_penalty": 1
    }
}

headers = {
    "Content-Type": "application/json"
}

response = requests.post(url, json=payload, headers=headers)

if response.status_code == 200:
    data = response.json()
    print("Response:", data.get("response", "No response key found"))
else:
    print("Request failed with status:", response.status_code)
    print(response.text)
```
